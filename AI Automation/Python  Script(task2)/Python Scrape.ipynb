```python
import requests
from bs4 import BeautifulSoup

def get_about_us(url):
    try:
        response = requests.get(url, timeout=10)
        soup = BeautifulSoup(response.text, 'html.parser')
        for tag in soup.find_all(['h1', 'h2', 'p', 'div']):
            if tag and 'about' in tag.get_text().lower():
                return tag.get_text().strip()
        return "'About Us' section not found."
    except Exception as e:
        return f"Error fetching {url}: {e}"

# check for other websites
websites = [
    "https://growify.in",     #Example url 1
    "https://apple.com"      #Example url 2
]

for url in websites:
    about_text = get_about_us(url)
    print(f"\n Website: {url}")
    print("About Us:\n", about_text)
```
